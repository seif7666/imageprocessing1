# In this task, you're asked to write a program that does the following
- Read video from camera
- crop a 400*400 box
- draw a blue 20*20 square in the upper left corner
- draw a red 20*20 square in the upper right corner

```python
#Read video
vid = cv2.VideoCapture(0)
read,pic = vid.read()
pic=pic[:400,:400]#Crop
pic=cv2.rectangle(pic , (0,0),(20,20),(255,0,0),-1)
pic=cv2.rectangle(pic , (400-20,0),(400,20),(0,0,255),-1)    
   # cv2.imshow(' ',pic)
   # plt.axes(False)
plt.axis(False)    
plt.imshow(pic)
vid.release()
cv2.destroyAllWindows()
```
